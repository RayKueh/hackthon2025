```python

def jacobian(x_0,x_1,bias):
#---modified 0403---------
    lenx = np.shape(x_0)[1]
    jaco = linalg.inv(x_0.T@x_0+bias*np.eye(lenx))@x_0.T@x_1
    return jaco


def jacobian_effective(x_0,x_1):
#---modified 0403---------#
    lenx = np.shape(x_0)[1]
    jaco = linalg.inv(x_1.T@x_0)@x_1.T@x_1
    return jaco

def foward_wide2(x_ini,jaco,steps,name=False):
    # for lorenz model , len(x_ini) = 3
    lenx = np.shape(x_ini)[0]
    output = np.zeros((steps,lenx))
    output[0,:] = x_ini
    if name == False:  
        for i in range(steps-1):
            temp        = widedata_2(output[i].reshape((1,lenx))).flatten()
            temp2       = np.hstack((output[i],temp))
            out_wide    = jaco@temp2
            output[i+1] = out_wide[:lenx]
        return output
    else: 
        for i in range(steps-1):
            temp        = widedata_2(output[i].reshape((1,lenx))).flatten()
            temp2       = np.hstack((output[i],temp))
            temp3       = np.delete(temp2,np.array(name), axis=0)
            out_wide    = jaco@temp3
            output[i+1] = out_wide[:lenx]
        return output
def widedata_2(a):
    a = np.array(a)
    dim = len(np.shape(a))
    time = np.shape(a)[0]    # length of time
    state = np.shape(a)[1]   # length of state vector
    
    # generate a bigger array to store tensored data
    t0 = np.ones(state)
    temp = np.kron(t0,t0)*np.nan
    state_big = np.shape(temp.flatten())[0]
    output = np.nan*np.zeros((time,state_big))
    #print(temp.shape)
    #print(output.shape)
    for i in range(time):
        state_now = a[i,:]
        temp0 = np.kron(state_now,state_now)
        size = temp0.shape[0]
        output[i,:size] = temp0
    #output2 = output[~(output==0)]
    #shape2 = int(len(output2)/time)
    return output
```

```python
def find_nearest(point,a,persent):
    # a : the array contains the point
    # a : (time, points)
    # persent : nearest 
    # return : indexs
    size      = a.shape[0]
    k         = int(size*persent)
    points    = np.tile(point,(size,1))
    norm      = np.sum((points-a)**2,axis=1)
    
    top_k_idx = norm.argsort()[::1][0:k]
    return top_k_idx

def det_real_data(data,persent,bias):  # data: 2D array
    size = data.shape[0]  # time variable
    num  = data.shape[1]  # space variable
    det  = np.zeros(size)
    for i in tqdm(range(size)): 
        p = data[i]
        idx = find_nearest(p,data,persent)
        idx = idx[idx+1<size]
        x_0 = data[idx,:]
        x_1 = data[idx+1,:]
        #----------------------A = torch.Tensor(x_0.T@x_0+bias*np.eye(num))
        #----------------------jaco   = torch.linalg.inv(A)@x_0.T@x_1
        A = torch.Tensor(x_1.T@x_0+bias*np.eye(num))
        jaco   = torch.linalg.inv(A)@x_1.T@x_1
        det[i] = torch.linalg.det(torch.Tensor(jaco)).item()
    return det

def foward(x_ini,jaco,steps):
    lenx = np.shape(x_ini)[0]
    output = np.zeros((steps,lenx))
    output[0,:] = x_ini
    for i in range(steps-1):
        output[i+1] = output[i]@jaco
    return output
```

```python
def det_real_data_2(data,data2,persent,bias):  # data: grids [??,2]    data2: original data ,[?,2]
    size = data.shape[0]  # time variable
    num  = data.shape[1]  # space variable
    det  = np.zeros(size)
    for i in tqdm(range(size)): 
        p = data[i]
        idx = find_nearest(p,data2,persent)
        idx = idx[idx+1<13514]
        x_0 = data2[idx,:]
        x_1 = data2[idx+1,:]
        A = torch.Tensor(x_0.T@x_0+bias*np.eye(num))
        jaco   = torch.linalg.inv(A)@x_0.T@x_1
        det[i] = torch.linalg.det(torch.Tensor(jaco)).item()
    return det
```

```python
def diff_angle(angle,ref_angle):
  #ref_angle = 225.0
  dim = len(angle.shape)
  
  if dim == 1:
        diff1     = np.abs(2*np.pi+angle-ref_angle)
        diff2     = np.abs(angle-ref_angle)
        temp      = np.vstack((diff1,diff2))
        return   np.min(temp,axis=0)
  if dim >= 2 : 
        diff1     = np.abs(2*np.pi+angle-ref_angle)
        diff2     = np.abs(angle-ref_angle)
        temp      = np.concatenate([diff1,diff2],axis=0)
        return   np.min(temp,axis=0)
```

```python
def gradient_central(f,x,h):
    num  =  len(x)
    grad =  np.zeros(num)
    for i in range(num): 
        x0       = np.copy(x)
        x1       = np.copy(x)
        x2       = np.copy(x)
        x3       = np.copy(x)
        
        x0[i]    = x0[i]+2*h
        x1[i]    = x1[i]+h
        x2[i]    = x2[i]-h
        x3[i]    = x3[i]-2*h
        
        grad[i]  = (-f(x0)+8*f(x1)-8*f(x2)+f(x3))/(12*h)
    return grad

#https://en.wikipedia.org/wiki/Gradient_descent
def grad_decent(guess,f,r):  # r : learning rate
    tol = 1000
    x   = np.copy(guess)
    iterate = 0
    v       = 0
    while(tol>1e-20 and iterate<2000):
        #----------------choose h--------------
        if tol >= 0.01 :
            h = 0.0001
        else:
            h=0.0001
        #-----------momentum--------------------
        if iterate == 0: 
            vnew = r*gradient_central(f,x,h)
            v    = 0 
        else:
            vnew = 0.9*v+r*gradient_central(f,x,h)
            v    = vnew
        #---------------------------------------
        xnew = x - vnew
        #xnew = x
        tol  = linalg.norm(xnew-x)*1e2
        x    = xnew
        iterate = iterate+1
        if iterate % 100 == 0 : 
            print(iterate, 'tol=',tol)
    return x
```

```python
import numpy as np

def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    phi2 = np.where(phi<=0,phi+2*np.pi,phi)
    
    return(rho, phi2)

def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)
```

```python
def gauss_grid(x0,num,f,r): # x: flattened array, num: num of gaussian kernel
    total1 = np.zeros_like(f)
    total2 = np.zeros_like(f)
    x     = x0.reshape((num,5))
    
    for i in range(num):
        fref   = x[i,0]
        rref   = x[i,1]
        sf     = x[i,2]
        sr     = x[i,3]
        c      = x[i,4]
        temp   = c*np.exp( 
                        -((f-fref)/sf)**2  -((r-rref)/sr)**2
                        )
        total1  = total1+temp
    f2 = -2*np.pi+f
    for i in range(num):
        fref   = x[i,0]
        rref   = x[i,1]
        sf     = x[i,2]
        sr     = x[i,3]
        c      = x[i,4]
        temp   = c*np.exp( 
                        -((f2-fref)/sf)**2  -((r-rref)/sr)**2
                        )
        total2  = total2+temp
    return total1+total2
```
